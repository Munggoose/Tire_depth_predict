```python
import os
from glob import glob
import random
import shutil
```

```python
root = 'F:\\data\Tire_data\\220920\\2차데이터_Masked\masked'
test_dir = 'F:\\data\Tire_data\\220920\\2차데이터_Masked\\test'
train_dir = 'F:\\data\Tire_data\\220920\\2차데이터_Masked\\train'
dirs = os.listdir(root)
```

```python
for dir in dirs:
    dir_path = os.path.join(root,dir)
    test_dir_path = os.path.join(test_dir,dir)
    train_dir_path = os.path.join(train_dir,dir)
    os.makedirs(test_dir_path,exist_ok=True)
    os.makedirs(train_dir_path,exist_ok=True)
    imgs = glob(os.path.join(dir_path,'*'))
    cnt = len(imgs)//4
    test_imgs = random.sample(imgs,cnt)
    train_imgs = list(set(imgs) - set(test_imgs))

    for test_i in test_imgs:
        f_name = os.path.basename(test_i)
        shutil.copy(test_i,os.path.join(test_dir_path,f_name))

    for train_i in train_imgs:
        f_name = os.path.basename(train_i)
        shutil.copy(train_i,os.path.join(train_dir_path,f_name))

```
